```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('twitter_validation.csv')
data
```

```python
number = data.groupby("company").Number.sum().values.tolist()
# number.sort()
number
```

```python
indexes = data.groupby("company").Number.sum().index
indexes
```

```python
plt.bar(indexes , number )
plt.xticks(rotation=90)
```

```python
plt.pie(data.Number.head() , labels = data.company.head() , autopct = "%1.1f%%")
```
