```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv('Netflix Userbase.csv')
```

```python
data
```

```python
data_Can = data[data.loc[:, "Country"] == "Canada"]
```

```python
data_Can.groupby('Subscription Type').Gender.value_counts()
```

```python
lbl = data_Can[data_Can.loc[: , "Subscription Type"] == "Basic"].Gender.value_counts().index
plt.pie(data_Can[data_Can.loc[: , "Subscription Type"] == "Basic"].Gender.value_counts() ,labels = lbl, autopct='%1.1f%%' )
plt.title("Canada")
plt.legend()
```

```python
datalbl = data.groupby("Country").Age.sum().index.tolist()
# plt.hist(data.groupby("Country").Age.mean(), bins=5, color='skyblue', edgecolor='black', alpha=0.8 , xlabel = datalbl )
plt.pie(data.groupby("Country").Age.mean() , labels = datalbl , autopct = "%1.1f%%")
plt.title("Mean Age of Every Country")
```

```python
sur = data.groupby("Subscription Type").Device.value_counts()

sur1 = sur.reset_index()
plt.bar(sur1['Subscription Type'], sur.values)
plt.title("No. of Devices used per Subcriptions" , fontsize = 16 )
plt.xlabel("Subcription type")
plt.ylabel("Number of Devices")
```

```python
sub = data.loc[: , "Subscription Type"].value_counts()
plt.bar(sub.index , sub.values)
plt.title("Subcriptions" , fontsize = 16 )
plt.xlabel("Subcription type")
plt.ylabel("Number of Users")
```

```python
newd = data.groupby("Country")["Monthly Revenue"].sum()
sns.barplot(y=newd.values, x=newd.index, palette="viridis")
plt.xticks(rotation = 45)
plt.title("Monthly Revenue by Country", fontsize=16)
plt.xlabel("Total Monthly Revenue", fontsize=12)
plt.ylabel("Country", fontsize=12)
```

```python
syb = data.groupby('Subscription Type')['Monthly Revenue'].sum()
plt.pie(syb.values , labels = syb.index , autopct = "%1.1f%%")
plt.title("Subcription Type by Monthly Revenue" , fontsize = 16)
plt.legend()
```

```python
sns.boxplot(x='Subscription Type', y='Monthly Revenue', data=data, palette="Set3")
plt.title('Monthly Revenue by Subscription Type')
```

```python
dev = data.Device.value_counts()
dev.plot(kind='bar', color='lightblue', title='Device Usage Analysis', figsize=(6, 5))
plt.xticks(rotation = 45)
```

```python
new = data.groupby(["Subscription Type" , "Device"])["Age Group"].apply(list)
```

```python
grouped_data = data.groupby(["Subscription Type", "Device"])["Age Group"].count()

grouped_data = grouped_data.reset_index()

plt.figure(figsize=(10, 6))
plt.bar(
    grouped_data["Subscription Type"] + " - " + grouped_data["Device"],  # Combine labels
    grouped_data["Age Group"]
)
plt.xlabel("Subscription Type - Device")
plt.ylabel("Count of Age Group")
plt.title("Counts of Age Groups by Subscription Type and Device")
plt.xticks(rotation=45)
plt.show()
```
